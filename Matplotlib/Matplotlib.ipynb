### Matplotlib example workflow

```python
# 1. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 2. prepare data
x = [1,2,3,4,5]
y = [10,5,20,10,30]

# 3. setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 4. plot data
ax.plot(x,y)

# 5. customize plot
ax.set(title = 'Simple Plot',
      xlabel = 'x-axis',
      ylabel = 'y-axis')

# 6. save and show 
fig.savefig('sample-plot.png')
```

### Two options for subplots

```python
## subplot option 1

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10,10))

#plot to each different axes

ax0.plot(x, x/2);

ax1.scatter(np.random.random(10), np.random.random(10));

ax2.bar(prices.keys(), prices.values());

ax3.hist(np.random.randn(1000));
```

```python
## subplot option 2

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (7,7))

ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(prices.keys(), prices.values());
ax[1,1].hist(np.random.randn(100));
```

Above method is pyplot method which is not ideal as its using same scale for each feature.

* when plotting something quickly, we can use pyplot method
* when plotting advance, we should use OO (object oriented) method

```python
df.head()
```

```python
over_50 = df[df.age > 50]; over_50
```

```python
len(over_50)
```

```python
# pyplot method

over_50.plot(x='age',y='chol',c='target',kind='scatter');  #c = colour

#this is doing the work but it doesn't look good
```

```python
## OO method mixed with pyplot method

fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(x='age', y='chol',c='target', kind='scatter',ax = ax);

# ax.set_xlim(50,60); # we can set the range of x-axis
```

```python
# OO method from scratch

fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

#Customize the plot
ax.set(title = 'Heart Disease and Cholestrol', xlabel = 'Age', ylabel = 'Cholestrol');

#Add Legends
ax.legend(*scatter.legend_elements(), title = 'Target')

# Add a horizontal line. lets say mean of cholestrol 
ax.axhline(over_50.chol.mean(), linestyle = '--')

# Add a vertical line. lets say mean of age
ax.axvline(over_50.age.mean(), linestyle = ':')
```

```python
## subplot of chol, thalach vs age
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize=(10,10), sharex = True)   #single figure , two axes

#add data to ax0
scatter = ax0.scatter(x = over_50.age, y = over_50.chol, c = over_50.target)

#customize ax0
ax0.set(title = 'Heart Disease and cholestrol levels', ylabel = 'Cholestrol')

#add legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'Target ')

#add mean line 
ax0.axhline(y = over_50.chol.mean(), linestyle = '--');



#add data to ax1
scatter = ax1.scatter(x = over_50.age, y = over_50.thalach, c = over_50.target)

#customize ax1
ax1.set(title='Heart Disease vs Thalach', xlabel = 'Age', ylabel = 'Max Heart Rate')

#add legend
ax1.legend(*scatter.legend_elements(), title = 'Target')

#add meanline
ax1.axhline(y = over_50.thalach.mean(), linestyle = '--');
```

```python
## changing the color scheme usning cmap ( winter, summer, )

plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(figsize = (10,6))

scatter = ax.scatter(x = over_50.age,
                     y = over_50.chol, 
                     c = over_50.target,
                     cmap = 'winter');  #this changes the color scheme

ax.set(title = 'Heart Disease wrt age and cholestrol', xlabel = 'Age', ylabel = 'cholestrol')

ax.legend(*scatter.legend_elements(), title = 'Target')

ax.axhline(over_50.chol.mean(), linestyle = '--', color = 'red')
ax.axvline(over_50.age.mean(), linestyle = ':', color = 'blue')
```

```python
#customizing the y and x axis limitations

fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10), sharex = True)

scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap='summer')
ax0.set(title = 'Heart Disease with cholestrol', ylabel = 'Cholestrol')
ax0.legend(*scatter.legend_elements(), title = 'Target')
ax0.axhline(over_50.chol.mean(), linestyle = '--')
#changing the x-axis limit
ax0.set_xlim([50,80])


scatter = ax1.scatter(x=over_50.age, y=over_50.thalach , c=over_50.target, cmap = 'winter')
ax1.set(title = 'Heart Disease with max heart beat', xlabel = 'Age', ylabel = 'Max heart beat')
ax1.legend(*scatter.legend_elements(), title = 'Target')
ax1.axhline(over_50.thalach.mean(), linestyle = '--')
#changing the y-axis limit
ax1.set_ylim([60, 200])

#adding title to the figure
fig.suptitle('Hear Disease Analysis', fontsize=16, fontweight = 'bold');
```
